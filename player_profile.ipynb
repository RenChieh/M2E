```python
#player basic profile in M2E-2
import numpy as np
import random
import pandas as pd
player_data = {"player": [],"age": [],"days/week": [],"run_type": [],"km/day": [],
        "shoes_type": [],"place": [],"mileage": [],"wear_rate_tolerate": [],"hurt_rate": []}

runtype = ["A","B","C","D","E"]
shoes_type=["Badminton","Running","Basketball"]
place=["street","park","playground","grassland","Forest"]

player=1
player_amount = 500
for i in range(player_amount):
  current_age = int(random.uniform(15,80))
  currnet_dayweek = int(random.uniform(1,7))
  current_runtype = random.choice(runtype)
  if current_runtype == "A":
    current_kmday = 0
    currnet_dayweek = 0
    hurt_rate=0
  elif current_runtype == "B":
    current_kmday = int(random.uniform(1,10))
  elif current_runtype == "C":
    current_kmday = int(random.uniform(11,20))
  elif current_runtype == "D":
    current_kmday = int(random.uniform(21,30))
  else:
    current_kmday = int(random.uniform(31,42))

  current_shoes_type = random.choice(shoes_type)
  current_place = random.choice(place)
  if current_shoes_type == "Badminton" and current_place == "street":
    current_mileage = int(random.uniform(300,500))
  elif current_shoes_type == "Badminton" and current_place == "park":
    current_mileage = int(random.uniform(300,500))
  elif current_shoes_type == "Badminton" and current_place == "playground":
    current_mileage = int(random.uniform(200,400))
  elif current_shoes_type == "Badminton" and current_place == "grassland":
    current_mileage = int(random.uniform(100,200))
  elif current_shoes_type == "Badminton" and current_place == "Forest":
    current_mileage = int(random.uniform(100,200))
  elif current_shoes_type == "Running" and current_place == "street":
    current_mileage = int(random.uniform(500,800))
  elif current_shoes_type == "Running" and current_place == "park":
    current_mileage = int(random.uniform(500,800))
  elif current_shoes_type == "Running" and current_place == "playground":
    current_mileage = int(random.uniform(300,500))
  elif current_shoes_type == "Running" and current_place == "grassland":
    current_mileage = int(random.uniform(200,400))
  elif current_shoes_type == "Running" and current_place == "Forest":
    current_mileage = int(random.uniform(200,400))
  elif current_shoes_type == "Basketball" and current_place == "street":
    current_mileage = int(random.uniform(500,800))
  elif current_shoes_type == "Basketball" and current_place == "park":
    current_mileage = int(random.uniform(500,800))
  elif current_shoes_type == "Basketball" and current_place == "playground":
    current_mileage = int(random.uniform(300,500))
  elif current_shoes_type == "Basketball" and current_place == "grassland":
    current_mileage = int(random.uniform(200,400))
  elif current_shoes_type == "Basketball" and current_place == "Forest":
    current_mileage = int(random.uniform(200,400))
  else:
    current_mileage = 0
  current_wear_rate_tolerate = random.uniform(0,1.5)
  if(current_runtype != "A"):
    if current_age > 60 :
      hurt_rate = random.uniform(0.8,1)
    elif 60>=current_age>50:
      hurt_rate = random.uniform(0.6,0.8)
    elif 50>=current_age>40 :
      hurt_rate = random.uniform(0.4,0.6)
    elif 40>=current_age>30 :
      hurt_rate = random.uniform(0.2,0.4)
    elif current_age<=30:
      hurt_rate = random.uniform(0,0.2)
    else:
      hurt_rate = 0

  player_data["player"].append(i+1)
  player_data["age"].append(current_age)
  player_data["days/week"].append(currnet_dayweek)
  player_data["run_type"].append(current_runtype)
  player_data["km/day"].append(current_kmday)
  player_data["shoes_type"].append(current_shoes_type)
  player_data["place"].append(current_place)
  player_data["mileage"].append(current_mileage)
  player_data["wear_rate_tolerate"].append(round(current_wear_rate_tolerate,2))
  player_data["hurt_rate"].append(round(hurt_rate,2))

# 整理成資料表
player_profile = pd.DataFrame(player_data)
player_profile
```

```python
#player basic profile in M2E-3
#player profile
import numpy as np
import random
import pandas as pd
player_data = {"player": [],"age": [],"days/week": [],"run_type": [],"km/day": [],
        "shoes_type": [],"place": [],"mileage": [],"wear_rate_tolerate": [],"hurt_rate": []}
shoes_type=["Badminton","Running","Basketball"]
place=["street","park","playground","grassland","Forest"]

# 定義runtype及其相應的機率
# 總人數

player=1
player_amount = 300
# 設定每種runtype的人數，E型人佔%
num_E = int(player_amount * 0.1)
num_others = player_amount - num_E
num_each_other_type = num_others // 4
# 創建包含所有人的runtype列表
runtype_list = (["E"] * num_E +
                ["A"] * num_each_other_type +
                ["B"] * num_each_other_type +
                ["C"] * num_each_other_type +
                ["D"] * num_each_other_type)
# 如果剩餘人數不是4的倍數，則隨機分配剩下的人
remaining = num_others % 4
other_types = ["A", "B", "C", "D"]
for i in range(remaining):
    runtype_list.append(other_types[i])
# 隨機打亂runtype列表
random.shuffle(runtype_list)
# #模擬所有資料
for i in range(player_amount):
  current_age = int(random.uniform(15,80))
  currnet_dayweek = int(random.uniform(1,7))
  # 隨機選擇一個runtype
  current_runtype = runtype_list[i]
  if current_runtype == "A":
    currnet_dayweek = 0
    current_kmday = 0
    hurt_rate=0
  elif current_runtype == "B":
    current_kmday = int(random.uniform(1,10))
  elif current_runtype == "C":
    current_kmday = int(random.uniform(11,20))
  elif current_runtype == "D":
    current_kmday = int(random.uniform(21,30))
  else:
    currnet_dayweek = 7
    current_kmday = 42

  current_shoes_type = random.choice(shoes_type)
  current_place = random.choice(place)
  if current_shoes_type == "Badminton" and current_place == "street":
    current_mileage = int(random.uniform(300,500))
  elif current_shoes_type == "Badminton" and current_place == "park":
    current_mileage = int(random.uniform(300,500))
  elif current_shoes_type == "Badminton" and current_place == "playground":
    current_mileage = int(random.uniform(200,400))
  elif current_shoes_type == "Badminton" and current_place == "grassland":
    current_mileage = int(random.uniform(100,200))
  elif current_shoes_type == "Badminton" and current_place == "Forest":
    current_mileage = int(random.uniform(100,200))
  elif current_shoes_type == "Running" and current_place == "street":
    current_mileage = int(random.uniform(500,800))
  elif current_shoes_type == "Running" and current_place == "park":
    current_mileage = int(random.uniform(500,800))
  elif current_shoes_type == "Running" and current_place == "playground":
    current_mileage = int(random.uniform(300,500))
  elif current_shoes_type == "Running" and current_place == "grassland":
    current_mileage = int(random.uniform(200,400))
  elif current_shoes_type == "Running" and current_place == "Forest":
    current_mileage = int(random.uniform(200,400))
  elif current_shoes_type == "Basketball" and current_place == "street":
    current_mileage = int(random.uniform(500,800))
  elif current_shoes_type == "Basketball" and current_place == "park":
    current_mileage = int(random.uniform(500,800))
  elif current_shoes_type == "Basketball" and current_place == "playground":
    current_mileage = int(random.uniform(300,500))
  elif current_shoes_type == "Basketball" and current_place == "grassland":
    current_mileage = int(random.uniform(200,400))
  elif current_shoes_type == "Basketball" and current_place == "Forest":
    current_mileage = int(random.uniform(200,400))
  else:
    current_mileage = 0
  current_wear_rate_tolerate = random.uniform(0,1.5)
  if(current_runtype != "A"):
    if current_age > 60 :
      hurt_rate = random.uniform(0.8,1)
    elif 60>=current_age>50:
      hurt_rate = random.uniform(0.6,0.8)
    elif 50>=current_age>40 :
      hurt_rate = random.uniform(0.4,0.6)
    elif 40>=current_age>30 :
      hurt_rate = random.uniform(0.2,0.4)
    elif current_age<=30:
      hurt_rate = random.uniform(0,0.2)
    else:
      hurt_rate = 0

  player_data["player"].append(i+1)
  player_data["age"].append(current_age)
  player_data["days/week"].append(currnet_dayweek)
  player_data["run_type"].append(current_runtype)
  player_data["km/day"].append(current_kmday)
  player_data["shoes_type"].append(current_shoes_type)
  player_data["place"].append(current_place)
  player_data["mileage"].append(current_mileage)
  player_data["wear_rate_tolerate"].append(round(current_wear_rate_tolerate,2))
  player_data["hurt_rate"].append(round(hurt_rate,2))

# 整理成資料表
player_profile10 = pd.DataFrame(player_data)
player_profile10
```
